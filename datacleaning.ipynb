```python
!pip install langdetect  
from langdetect import detect
import pandas as pd
```

```python
df_review=pd.read_csv("/Users/yingzhaozhou/reviews.csv")

df_review.dropna()
```

```python
def trans(text):
    textresult=type(text).__name__
    return(textresult)
```

```python
df_review["type"]=df_review.comments.apply(trans)
```

```python
df_review
df_review=df_review[df_review["type"]=="str"]
df_review
```

```python
df_review["len"]=df_review.comments.apply(len)
df_review=df_review[df_review['len']>10]
df_review
```

```python
df_review[["listing_id","comments"]]
df_review["detect"]=df_review.comments.apply(detect)
```

```python
df_review=df_review[df_review["detect"]=="en"]
df_review
```

```python
df_review.to_csv("review_complete.csv")
```
